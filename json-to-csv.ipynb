```python
for json_path in to_remove:
    try:
        with open(json_path, 'r', encoding='utf-8') as file:
                data = json.load(file)
                try:
                    imgs = data['resource_response']['data']['results']
                except Exception as e:
                    raise e
                
                if len(imgs) == 1 and 'objects' in imgs[0]:
                    imgs = imgs[0]['objects']
                                
                for ind, img in enumerate(imgs):
                    total+=1
                    
                    if 'display_options' in img:
                        continue
                    
                    try:
                        id = img['node_id']
                    except Exception as e:
                        raise e
                    
                    try:
                        auto_alt_text = img['auto_alt_text']
                    except:
                        auto_alt_text = None
                    
                    try:
                        grid_title = img['grid_title']
                    except:
                        grid_title = None
                    
                    try:
                        description = img['description']
                    except:
                        description = None
                    
                    try:
                        grid_title = img['grid_title']
                    except:
                        grid_title = None
                    
                    try:
                        urls = img['images']
                    except:
                        try:
                            urls = img['cover_images']
                        except:
                            urls = None
                    
                    image = {
                        "id": id,
                        "title": img['title'],
                        "grid_title": grid_title,
                        "auto_alt_text": auto_alt_text,
                        "description": description,
                        "urls": urls,
                        "category": category
                    }
                    images.append(image)
                    
        passed.append(json_path)
    except RemoveExcepion as _:
        # os.remove(json_path)
        print("RemoveExcepion")
        print(json_path)
        pass
    except Exception as e:
        print(json_path)
        print(ind)
        print(img)
        raise e
```

```python
import csv

# Define the output CSV file
output_csv = 'cartoon_dataset.csv'

flattened_data = []
for item in images:
    try:
        urls = item['urls'].items()
    except:
        urls = item['urls'][0].items()
    
    urls_flat = {f"url_{size}": details['url'] for size, details in urls}
    flattened_row = {**item, **urls_flat}
    flattened_row.pop('urls', None)
    flattened_data.append(flattened_row)

# Get all column names (keys) dynamically
columns = set()
for row in flattened_data:
    columns.update(row.keys())
columns = sorted(columns)

# Write to CSV
with open(output_csv, 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=columns)
    writer.writeheader()
    writer.writerows(flattened_data)
```
